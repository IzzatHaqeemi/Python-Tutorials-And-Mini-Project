*Ignore any warning if given by the Python*


```python
# WRITE YOUR CODE HERE
tallest.index=list(range(0,tallest.shape[0]))
missing_Height=[]
Height=tallest.loc[:,'Height (meters)']
Category=tallest.loc[:,'Category']
bool_Height=Height.isnull()
n=-1
for i in bool_Height:
    n=n+1
    if(i==True):
        missing_Height.append(Category[n])
        tallest.loc[n,'Height (meters)']=tallest.loc[n,'Height (feet)']*0.3
print(missing_Height)
tallest
```

(i)	Create a function named ``eliminate_ref``. This function will receive an array and perform these tasks:

    (i)  Find whether there is any reference attached to any string in the array given. Commonly, any references can be detected by a square bracket containing a number like [5] as shown in figure below.
    
<br>
<img src="Figure 7.png", style="width: 400px";>

    (ii) If there is any reference attached, replace the word with the original word but without the reference. Set inplace=True when performing the replace method.

Test your function using the codes given below.

```python
# WRITE YOUR CODE HERE
import re
def eliminate_ref(arr):
    for i in range(len(arr)):
        ret = re.search(r'\[\d+\]', arr[i])
        if ret:
            substitution = re.sub(r'\[\d+\]', '', arr[i])
            arr[i] = substitution
```

```python
# TEST YOUR CODE HERE

eliminate_ref(tallest['Category'])
eliminate_ref(tallest['Structure'])
tallest
```
