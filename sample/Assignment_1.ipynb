### Question 6

A list of names in the format "first_name last_name" is given. Your task is to sort the list of names based on the last name in ascending order. Write a function called `sort_names_by_last_name` that takes in a list of names as input and uses a lambda function along with the ``sorted()`` function to solve the problem.

Example:<br>
If the input list is <br> `['John Smith', 'Alice Johnson', 'Michael Brown', 'Emily Davis']`, <br> the function should return <br> `['Michael Brown', 'Emily Davis', 'Alice Johnson', 'John Smith']`.

````python
#Write your code here

def sort_name_by_last_name(names):

  sort_name = sorted(names, key = lambda x : x.split()[-1])
  return sort_name

input_names = ['John Smith', 'Alice Johnson', 'Michael Brown', 'Emily Davis']
sort_name = sort_name_by_last_name(input_names)
print(sort_name)
````

If the input list is <br> `['apple', 'banana', 'cherry', 'date']`,<br> the function should return <br>`{'apple': 5, 'banana': 6, 'cherry': 6, 'date': 4}`.

````python
def string_lengths_dict():
    n = int(input("Enter the number of fruits: "))
    fruit_list = []
    FruitDict = {}

    for i in range(n):
        fruit = input("Enter a fruit: ")
        fruit_list.append(fruit)

    print("List of fruits:", fruit_list)

    for i in range (n):
      FruitDict[fruit_list[i]] = len(fruit_list[i])

    print("The Key and Value are: " , FruitDict)

string_lengths_dict()
````

````python
import pandas as pd
column_names = ["Date", "Item", "Quantity", "Price"]
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DATASET/Data_1.txt', sep=' ', names=column_names)
df
````

````python
def calculate_daily_sale(dataframe):

    daily_sales = {}


    for index, row in dataframe.iterrows():
        date = row['Date']
        item = row['Item']
        quantity = row['Quantity']
        price = row['Price']
        daily_sale = quantity * price


        if date not in daily_sales:
            daily_sales[date] = {}


        if item not in daily_sales[date]:
            daily_sales[date][item] = 0

        daily_sales[date][item] += daily_sale

    return daily_sales


result = calculate_daily_sale(df)


print(result["2020-09-25"])
print(result["2020-09-25"]["Item_B"])

````

````python
def get_total_sale(daily_sales):
    total_sales = {}

    for date, items in daily_sales.items():
        for item, sale in items.items():
            if item not in total_sales:
                total_sales[item] = 0
            total_sales[item] += sale

    return total_sales


result = calculate_daily_sale(df)

total_sales = get_total_sale(result)

print("Total Sales For Item_A is :" ,total_sales["Item_A"])
print("Total Sales For Item_B is :" ,total_sales["Item_B"])
print("Total Sales For Item_C is :" ,total_sales["Item_C"])
print("Total Sales For Item_D is :" ,total_sales["Item_D"])
print("Total Sales For Item_E is :" ,total_sales["Item_E"])
print("Total Sales For Item_F is :" ,total_sales["Item_F"])
````
