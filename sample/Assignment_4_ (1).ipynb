(d)	The products in the sales data can be further grouped as follow:

    a.Category A: facecream, facewash, moisturizer
    b.Category B: toothpaste
    c.Category C: bathingsoap, shampoo
    
    
Group the data accordingly and obtain the following information, then plot into two subplots:

    1.total unit sold of each category for every month
    2.percentage of total unit sold of each category for every month over the grand total of unit sold of that category.


```python
sales['Category A'] = sales[['facecream', 'facewash', 'moisturizer']].sum(axis=1)
sales['Category B'] = sales['toothpaste']
sales['Category C'] = sales[['bathingsoap', 'shampoo']].sum(axis=1)

total_units = sales[['Category A', 'Category B', 'Category C']].sum()
sales[['% Category A', '% Category B', '% Category C']] = sales[['Category A', 'Category B', 'Category C']].div(total_units) * 100

fig, axs = plt.subplots(2, figsize=(12, 12))

sales[['Category A', 'Category B', 'Category C']].plot(ax=axs[0])
axs[0].set_title('Total Units Sold for Each Category')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Units Sold')

sales[['% Category A', '% Category B', '% Category C']].plot(ax=axs[1])
axs[1].set_title('Percentage of Total Units Sold for Each Category')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Percentage (%)')

plt.show()
```

### Question 4

(a)	The file named `Expenditure.xlsx` describes the expenditure (in dollars) on recreation per month by employees at a certain company, and their corresponding monthly incomes.

    (i)	Using simple linear regression method, find the equation of the regression line.
    (ii)	Find the slope and interception values for the regression line.
    (iii)   Plot the data with the regression line.
    (iv)	Then estimate the monthly income of an employee at this company who spends 5000 dollars per month on recreation.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_excel(r'/content/Expenditure.xlsx')
data.head()
```

**(i) (ii) (iii)**

As we could see from the acquire, we need to use the simple linear regression method - the formula **y=kx+b** to regress.

```python

expenditure=data['Expenditure']
income=data['Income']

# (i) (ii)
slope, intercept = np.polyfit(expenditure, income, 1)
print(f"Regression line equation: y = {slope:.2f}x + {intercept:.2f}")
print("slope=%.2f,intercept=%.2f"%(slope,intercept))

#（iii）
plt.scatter(expenditure, income, label='Data')
plt.plot(expenditure, slope * expenditure + intercept, color='red', label='Regression Line')

plt.xlabel('Expenditure')
plt.ylabel('Income')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

**(iv)**

```python
monthly_expense = 5000
estimated_income = slope * monthly_expense + intercept
print(f"the monthly income of an employee at this company who spends 5000 dollars per month on recreation is:\n{estimated_income:.2f}")
```

(i)	Load the `Titanic.csv` into a DataFrame named `Titanic`. Below is the data dictionary for the dataset.

<br>
<img src="Fig4.png", style="width: 500px";>

(ii) Perform dimensionality reduction to the dataset using the Principal Component Analysis (PCA) and next apply k-means clustering to the data.

**(i)**

```python
Titanic=pd.read_csv(r'/content//Titanic.csv')
Titanic.head()
```

**(ii)**

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# data cleaning
Titanic = Titanic.dropna()
# features
features = ["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]
X = Titanic[features]

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
Titanic["cluster"] = kmeans.fit_predict(X)

# visualization
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=Titanic["cluster"], cmap="viridis")
plt.title("The PCA and K-Means Clustering of Titanic Data")
plt.grid(color='0.8',linewidth='0.5',axis='y')
plt.xlabel("main1")
plt.ylabel("main2")
plt.show()
```
