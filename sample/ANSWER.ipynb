### Question 2(d) - (6 marks)

(i) Create a function that
   - read the ``sal_zip`` tuple
   - get the currency exchange from the ``ext_dict`` dictionary
   - calculate the currency conversion and round it to 2 decimal point
   - append the result into a list named ``calc_usd``

```python
def calculate_usd(sal_zip, ext_dict):
    calc_usd = []
    for salary_currency, salary_amount in sal_zip:
        if salary_currency in ext_dict:
            currency_exchange = ext_dict[salary_currency]
            converted_amount = round(salary_amount * currency_exchange, 2)
            calc_usd.append(converted_amount)
        else:
            calc_usd.append(None)
    return calc_usd


result = calculate_usd(sal_zip, exc_dict)
print(result)
```

### Question 3(d) - (16 marks)

Group the math, reading and writing score by gender. Find the average of score for each gender fo each group and produce a bar plot for each result as shown below. Set the figure size as (8,8).

<br>
<img src="Figure 6.png", style="width: 500px";>

```python
# write your code here
grouped = result.groupby('gender')[['math score', 'reading score', 'writing score']]
mean_scores = grouped.mean()

math = mean_scores[['math score']]
print(math)

reading = mean_scores[['reading score']]
print(reading)

writing = mean_scores[['writing score']]
print(writing)
```

```python
fig,axs = plt.subplots(1,3,figsize=(8,8))

math.plot (ax = axs[0],kind='bar',color='blue')
axs[0].set_title('Math Score by Gender')
axs[0].set_ylabel('Score')

reading.plot (ax = axs[1],kind='bar',color='purple')
axs[1].set_title('Reading Score by Gender')
axs[1].set_ylabel('Score')

writing.plot (ax = axs[2],kind='bar',color='green')
axs[2].set_title('Writing Score by Gender')
axs[2].set_ylabel('Score')

plt.tight_layout()

plt.show()

```
