```python
math_score = result.groupby('gender')['math score'].mean()
reading_score = result.groupby('gender')['reading score'].mean()
writing_score = result.groupby('gender')['writing score'].mean()

fig , axes = plt.subplots(1,3 , figsize = (8,8))

math_score.plot (ax = axes[0] , kind='bar' , color='blue')
axes[0].set_title('Math Score By Gender')
axes[0].set_ylabel('Score')

reading_score.plot (ax = axes[1], kind='bar' , color='purple')
axes[1].set_title('Reading Score By Gender')
axes[1].set_ylabel('Score')

writing_score.plot (ax = axes[2], kind='bar' , color='green')
axes[2].set_title('Writing Score By Gender')
axes[2].set_ylabel('Score')

plt.tight_layout()


plt.show()
```
