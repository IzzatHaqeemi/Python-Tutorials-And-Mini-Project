a) Create a function name verbing which receive a string as the parameter. Upon receiving a
string: (7 marks)<br>
• if its length is at least 3: add 'ing' to its end.<br>
• Unless it already ends in 'ing', in which case add 'ly' instead.<br>
• If the string length is less than 3, leave it unchanged.<br>
• Return the resulting string.

```python
def verbing(sample):
    if len(sample) >= 3:
        if 'ing' not in sample:
            sample = sample + 'ing'
        else:
            sample = sample + 'ly'
    return sample


verbing('swimming')
```

c)Using regular expression, write the python code that can find all the network IP address and
replace it to 179.01.10.1.

```python
import re

paragraph = """
On most computer systems, localhost resolves to the IP address 10.100.11.121, which is the most commonly used IPv4 loopback address, and to the IPv6 loopback address. The localhost IP address is 192.168.11.10.
"""

ipv4_pattern = r'\b(?:\d{1,3}\.){3}\d{1,3}\b'

ipv4_addresses = re.findall(ipv4_pattern, paragraph)

new_paragraph = re.sub(ipv4_pattern, "179.01.10.1", paragraph)

print(new_paragraph)
```

```python
def eliminate_ref(arr, inplace=True):
    result = [i.split('[')[0] for i in arr]
    if inplace:
        arr[:] = result
    else:
        return result
```

```python
eliminate_ref(tallest['Category'])
eliminate_ref(tallest['Structure'])
tallest
```
